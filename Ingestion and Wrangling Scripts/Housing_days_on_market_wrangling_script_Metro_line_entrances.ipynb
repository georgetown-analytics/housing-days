# Housing Days On Market Wrangling For Metro line entrances Data

## Python Package(s) Used

```python
#import datetime as dt
#import dill
#import json
import numpy as np
#import os
import pandas as pd
#import requests
#import time
```

## Data Fetching

```python
# Read the entrance data into DataFrames
df_blue = pd.read_excel('BlueLIne_Entrance.xlsx')
df_green = pd.read_excel('GreenLine_Entrance.xlsx')
df_orange = pd.read_excel('OrangeLine_Entrance.xlsx')
df_red = pd.read_excel('RedLine_Entrance.xlsx')
df_silver = pd.read_excel('SilverLine_Entrance.xlsx')
df_yellow = pd.read_excel('YellowLine_Entrance.xlsx')
```

## Data Wrangling

```python
df_blue.head()
```

```python
# Rename columns to simplier names
df_blue = df_blue.rename(columns = lambda x:str(x)[4:])
df_green = df_green.rename(columns = lambda x:str(x)[4:])
df_orange = df_orange.rename(columns = lambda x:str(x)[4:])
df_red = df_red.rename(columns = lambda x:str(x)[4:])
df_silver = df_silver.rename(columns = lambda x:str(x)[4:])
df_yellow = df_yellow.rename(columns = lambda x:str(x)[4:])
```

```python
df_blue.head()
```

```python
df_metro_entrances = pd.concat([df_blue, df_green, df_orange,
                                df_red, df_silver, df_yellow], ignore_index = True)
```

```python
df_metro_entrances.to_csv('df_metro_entrances.csv')
```
